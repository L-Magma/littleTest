```python
from kubernetes import client, config
config.kube_config.load_kube_config()
v1 = client.CoreV1Api()
```

```python
for ns in v1.list_namespace().items:
    print(ns.metadata.name)
```

```python
#列出所有的pod
ret = v1.list_pod_for_all_namespaces(watch=False)
for i in ret.items:
    print("%s\t%s\t%s" % (i.status.pod_ip, i.metadata.namespace, i.metadata.name))
```

```python
ret = v1.list_service_for_all_namespaces(watch=False)
for i in ret.items:
    print("%s \t%s \t%s \t%s \t%s \n" % (i.kind, i.metadata.namespace, i.metadata.name, i.spec.cluster_ip, i.spec.ports ))
```

```python
for i in client.AppsV1beta2Api().list_deployment_for_all_namespaces(watch=False).items:
    print(i.metadata.namespace, i.metadata.name)
    break
```

```python
ret = client.AppsV1Api().list_deployment_for_all_namespaces_with_http_info(watch=False)
print(ret)
for i in ret.items:
    print("%s\t%s\t%s" % (i.status.pod_ip, i.metadata.namespace, i.metadata.name))
```
